# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Ejercicio 1

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Función de densidad de probabilidad (PDF)
def dnorm(x, mean=0, std=1):
    """
    Calcula la densidad de probabilidad de la distribución normal.
    """
    result = norm.pdf(x, loc=mean, scale=std)
    return result

# Función de distribución acumulada (CDF)
def pnorm(q, mean=0, std=1):
    """
    Calcula el acumulado de la distribución normal.
    """
    result = norm.cdf(q, loc=mean, scale=std)
    return result

# Función cuantil (PPF)
def qnorm(p, mean=0, std=1):
    """
    Calcula la función cuantil a partir de la distribución normal.
    """
    result = norm.ppf(p, loc=mean, scale=std)
    return result

# Generación de variables aleatorias (RVS)
def rnorm(n, mean=0, std=1):
    """
    Genera variables aleatorias a partir de la distribución normal.
    """
    result = norm.rvs(loc=mean, scale=std, size=n)
    return result
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Parámetros de la distribución normal
mean = 0
std = 1

print(f"Probability density at x=0 for N(0,1): {dnorm(0, mean=mean, std=std)}")
print(f"Cumulative probability at x=0 for N(0,1): {pnorm(0, mean=mean, std=std)}")
print(f"Quantile at p=0.9 for N(0,1): {qnorm(0.9, mean=mean, std=std)}")
print(f"Generate 10 random values from N(0,1): {rnorm(10, mean=mean, std=std)}")

# Gráficas
data = rnorm(1000, mean=mean, std=std)
pdf = dnorm(data, mean=mean, std=std)
cdf = pnorm(data, mean=mean, std=std)
ppf = qnorm(np.linspace(0, 1, 1000), mean=mean, std=std)

fig, axis = plt.subplots(3, 1, figsize=(5, 7))

axis[0].plot(data, pdf, 'r.')
axis[0].set_title('Probability Density Function (PDF)')

axis[1].plot(data, cdf, 'b.')
axis[1].set_title('Cumulative Distribution Function (CDF)')

axis[2].plot(np.linspace(0, 1, 1000), ppf, 'g.')
axis[2].set_title('Percent Point Function (PPF)')

plt.tight_layout()
plt.show()
```

```python
#distribucion uniforme

from scipy.stats import uniform

# Función de densidad de probabilidad (PDF)
def dunif(x, start=0, width=1):
    """
    Calcula la densidad de probabilidad de la distribución uniforme.
    """
    result = uniform.pdf(x, loc=start, scale=width)
    return result

# Función de distribución acumulada (CDF)
def punif(q, start=0, width=1):
    
    #Calcula el acumulado de la distribución uniforme.
    result = uniform.cdf(q, loc=start, scale=width)
    return result

# Función cuantil (PPF)
def qunif(p, start=0, width=1):
    
    #Calcula la función cuantil a partir de la distribución uniforme.
    
    result = uniform.ppf(p, loc=start, scale=width)
    return result

# Generación de variables aleatorias (RVS)
def runif(n, start=0, width=1):
    
    #Genera variables aleatorias a partir de la distribución uniforme.
    result = uniform.rvs(loc=start, scale=width, size=n)
    return result
```

```python
#aplicamos ejemplo
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Parámetros de la distribución uniforme
start = 0
width = 1

print(f"Probability density at x=0.5 for U(0,1): {dunif(0.5, start=start, width=width)}")
print(f"Cumulative probability at x=0.5 for U(0,1): {punif(0.5, start=start, width=width)}")
print(f"Quantile at p=0.9 for U(0,1): {qunif(0.9, start=start, width=width)}")
print(f"Generate 10 random values from U(0,1): {runif(10, start=start, width=width)}")

# Gráficas
data = np.arange(-5, 5, 0.01)
pdf = dunif(data, start=start, width=width)
cdf = punif(data, start=start, width=width)
ppf = qunif(np.linspace(0, 1, 1000), start=start, width=width)

fig, axis = plt.subplots(3, 1, figsize=(5, 7))

axis[0].plot(data, pdf, 'r-')
axis[0].set_title('Probability Density Function (PDF)')

axis[1].plot(data, cdf, 'b-')
axis[1].set_title('Cumulative Distribution Function (CDF)')

axis[2].plot(np.linspace(0, 1, 1000), ppf, 'g-')
axis[2].set_title('Percent Point Function (PPF)')

plt.tight_layout()
plt.show()
```
